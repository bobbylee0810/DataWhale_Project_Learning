# 异常检测—基于统计学的方法

## 参数方法  
> 假定正常的数据对象被一个以$\Theta$为参数的参数分布产生。该参数分布的概率密度函数$f(x,\Theta)$给出对象$x$被该分布产生的概率。该值越小，$x$越可能是异常点。

![一元正太分布](../pic/2.1.png)

1. 基于正态分布的一元异常点检测
    - $x^{(i)}\sim N(\mu, \sigma^2)$  
    - $p(x)=\frac 1{\sqrt{2\pi}\sigma}exp(-\frac{(x-\mu)^2}{2\sigma^2})$
    - 黄色线条是N(o,1)标准正态分布的概率密度函数图，我们看到，越靠近均值0的部分，概率密度越大，说明数据落在此点范围内的可能性更大
    - 蓝色线条是N(0,0.5)分布，可以看到，分布方差$\sigma$越大，数据集中程度越大
    - 参数方法，就是先假设原始数据，服从一个分布函数，然后通过最大似然估计等方法，通过实际样本的数据，计算出这个分布的参数($\mu,\sigma$)，从而给定新数据的时候，可以根据通过分布函数p(x)，计算出一个概率值，从而判断新数据是否是异常样本
2. 分量彼此独立的多元分布
    - $p(x)=\prod_{j=1}^n p(x_j;\mu_j,\sigma_j^2)=\prod_{j=1}^n\frac 1{\sqrt{2\pi}\sigma_j}exp(-\frac{(x_j-\mu_j)^2}{2\sigma_j^2})$

3. 分量存在相关关系的多元高斯分布
    - $p(x)=\frac{1}{(2 \pi)^{\frac{n}{2}}|\Sigma|^{\frac{1}{2}}} \exp \left(-\frac{1}{2}(x-\mu)^{T} \Sigma^{-1}(x-\mu)\right)$

## 非参数方法  
> 非参数方法，就是不假设数据服从某个分布，而是通过样本的实际分布来作为整体的分布，然后当给定新样本后，我们根据此分布，来判断新样本的出现概率，并根据此概率值，决策此样本是否是异常样本

<img src="../pic/2.2jpg" width = 50% alt="图片名称" align=center />  

- 如上图的直方图所示，可以看到直方图的高度越高，说明在此区间内的样本量越大，因此可以将方块的面积，作为概率密度，面积的倒数就可以作为异常度  
- 新样本落在面积比较小的方块内，异常度就越大
- **难点**就在于方块（箱）的数量选择，太多，则异常点易落入正常范围，太少，则正常值易误判为异常值。

## pyod中hbos函数简介  
```python
classpyod.models.hbos.HBOS(n_bins=10, alpha=0.1, tol=0.5, contamination=0.1)
```
- 参数介绍
    - n_bins 分区也就是分箱的数量  
    - alpha 正则化参数
    - tol 决定样本落在箱外概率的一个参数
    - contamination 数据集的异常值比例
- 属性
    - bin_edges_ 箱子的划分边界
    - hist_ 直方图的密度
    - 其他参数同一般模型

## hbos实例

### 生成数据

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.font_manager

from pyod.utils.data import generate_data,get_outliers_inliers

# 生成二维随机数据
X_train, Y_train = generate_data(n_train=200,train_only=True, n_features=2)

# 拆分出异常数据和正常数据
x_outliers, x_inliers = get_outliers_inliers(X_train,Y_train)

# 绘制生成的数据图
F1 = X_train[:,[0]].reshape(-1,1)
F2 = X_train[:,[1]].reshape(-1,1)
plt.scatter(F1,F2)
plt.xlabel('F1')
plt.ylabel('F2') 
plt.show()
````

### 使用HBOS进行预测

````python
from pyod.models.hbos import HBOS

outlier_fraction = 0.1#异常值比例
abod = HBOS(contamination=outlier_fraction)#生成模型
abod.fit(X_train)#训练模型
score = abod.decision_scores_# 注意异常得分是负的
score = score * -1
score.shape
````

### 预测结果

````python
y_pred = abod.predict(X_train)# 预测训练样本的标签
from sklearn.metrics import classification_report
print(classification_report(y_true=Y_train,y_pred=y_pred))
````

### 生成热力图

````python
n_inliers = len(x_inliers)
n_outliers = len(x_outliers)
#生成热力图的坐标点
xx , yy = np.meshgrid(np.linspace(-10, 10, 200), np.linspace(-10, 10, 200))
# 根据百分比生成
threshold = -abod.threshold_

# 得到每个坐标点的异常值得分
Z = abod.decision_function(np.c_[xx.ravel(), yy.ravel()]) * -1
Z = Z.reshape(xx.shape)
plt.figure(figsize=(10, 10))
#将正常样本区域绘制成蓝色
plt.contourf(xx, yy, Z, levels = np.linspace(Z.min(), threshold, 10),cmap=plt.cm.Blues_r)
#绘制决策曲线
a = plt.contour(xx, yy, Z, levels=[threshold],linewidths=2, colors='red')
#将异常样本区域绘制成橘黄色
plt.contourf(xx, yy, Z, levels=[threshold, Z.max()],colors='orange')
#绘制正常点（白色）
b = plt.scatter(X_train[:-n_outliers, 0], X_train[:-n_outliers, 1], c='white',s=20, edgecolor='k') 
# 绘制异常点（黑色）
c = plt.scatter(X_train[-n_outliers:, 0], X_train[-n_outliers:, 1], c='black',s=20, edgecolor='k')
plt.axis('tight')

plt.legend(
    [a.collections[0], b, c],
    ['learned decision function', 'true inliers', 'true outliers'],
    prop=matplotlib.font_manager.FontProperties(size=10),
    loc='lower right')

plt.title('HBOS')
plt.xlim((-10, 10))
plt.ylim((-10, 10))
plt.show()
````
